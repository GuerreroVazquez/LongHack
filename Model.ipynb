```python
import pandas as pd
```

```python
df = pd.read_csv('Enis_Age.csv')
```

```python
bins = [20, 35, 50, 100]
names = ['20-35', '35-50', '50+']

df['Age'] = pd.cut(df['Age'], bins, labels=names)
```

```python
from pycaret.classification import *
```

```python
exp_mclf101 = setup(data = df, normalize=True, target = 'Age', train_size=0.8, session_id=123) 
```

```python
best = compare_models(sort='F1')
```

```python
best_model = create_model(best)
```

```python
tuned_best = tune_model(best_model)
```

```python
plot_model(tuned_best, plot = 'confusion_matrix')
```

```python
evaluate_model(tuned_best)
```

```python
predict_model(tuned_best)
```
