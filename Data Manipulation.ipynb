```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Original_data/GSE164471_GESTALT_Muscle_ENST_TPM_annotated.csv')
```

```python
df.dropna(inplace=True)
```

```python
df = df.rename(columns={'HG19en82 Gene Name': 'Gene Name'})
```

```python
df = df[[column for column in df.columns if not 'HG19' in column]]
```

```python
df = df.T.reset_index()
```

```python
df.columns = df.iloc[-1]
df = df.iloc[:-1]
```

```python
def age_determine(x):
    return int(''.join([s for s in x.split('_')[1] if s.isdigit()]))
```

```python
df['Gene Name'] = df['Gene Name'].iloc[1:].map(age_determine)
```

```python
df = df.iloc[1:]
```

```python
df.rename(columns={'Gene Name':'Age'}, inplace=True)
```

```python
df.reset_index(inplace=True, drop=True)
```
